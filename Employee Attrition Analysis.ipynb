```python
df=data.copy()
```

```python
df.drop(['Over18','StandardHours','EmployeeNumber','EmployeeCount'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df[['RelationshipSatisfaction','JobSatisfaction','EnvironmentSatisfaction','JobInvolvement']]
```

```python
df['TotalSatisfaction_mean']=(df['RelationshipSatisfaction']+df['JobSatisfaction']+df['EnvironmentSatisfaction']+df['JobInvolvement']+df['WorkLifeBalance'])/5
```

```python
df['TotalSatisfaction_mean']
```

```python
# Create a new columns based on some condition
def satif(df):
    if df['TotalSatisfaction_mean']>2.35:
        return 1
    else:
        return 0
```

```python
df['Overall_satisf']=df.apply(lambda df:satif(df),axis=1)
df['Overall_satisf']
```

```python
def satif(df):
    if df['NumCompaniesWorked']>4:
        return 1
    else:
        return 0
df['MovingPeople']=df.apply(lambda df:satif(df),axis=1)
df['MovingPeople']
```
